```python
# Esse exercÃ­cio nos introduz a biblioteca "folium" onde Ã© possÃ­vel realizar tarefas de geolocalizaÃ§Ã£o 
import folium 

# Definindo as coordenadas de Porto Alegre 
latitudePOA = -30.0346 
longitudePOA = -51.2177 

# Definindo as coordenadas de SÃ£o Paulo 
latitudeSP = -23.5505 
longitudeSP = -46.6333 

# Definindo as coordenadas de Rio de Janeiro 
latitudeRJ = -22.9068 
longitudeRJ = -43.1729 

# Definindo as coordenadas de Curitiba 
latitudeCWB = -25.4290 
longitudeCWB = -49.2671 

# Definindo as coordenadas de Belo Horizonte 
latitudeBH = -19.9173 
longitudeBH = -43.9345 

# Criando um mapa centrado na mÃ©dia da latitude e longitude de ambas cidades
media_latitude = (latitudePOA + latitudeSP + latitudeRJ + latitudeCWB + latitudeBH) / 5
media_longitude = (longitudePOA + longitudeSP + longitudeRJ + longitudeCWB + longitudeBH) / 5
mapa = folium.Map(location=[media_latitude, media_longitude], zoom_start=6) 

# Fazendo marcaÃ§Ãµes nas cidades
folium.Marker ([latitudePOA, longitudePOA], popup='Porto Alegre, RS'). add_to(mapa) 
folium.Marker([latitudeSP, longitudeSP], popup='SÃ£o Paulo, SP').add_to (mapa) 
folium.Marker([latitudeRJ, longitudeRJ], popup='Rio de Janeiro, RJ').add_to (mapa) 
folium.Marker([latitudeCWB, longitudeCWB], popup='Curitiba, PR').add_to (mapa) 
folium. Marker([latitudeBH, longitudeBH], popup='Belo Horizonte, MG').add_to (mapa)

# Exibindo o mapa 
mapa
```
