```python
#Exercício para fazer uma análise de preferência pessoal sobre dataset
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("../games.csv")

top_10_games = df.nlargest(10, "Global_Sales")[["Name", "Global_Sales"]]

plt.figure(figsize=(16, 6))
plt.barh(top_10_games["Name"], top_10_games["Global_Sales"], color="blue")
plt.xlabel("Vendas no mundo (em milhões)")
plt.ylabel("Jogos")
plt.title("Top 10 Jogos Mais Vendidos de Todos os Tempos")
plt.gca().invert_yaxis()
plt.show()
```
