```python
#Exerc√≠cio que envolvia a mistura de gr√°fico com dataframe
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

dados = {
    'NOME':['Pedro','Maria','Janaina','Wong','Roberto','Marco','Paula'],
    'CIDADE':['Porto Alegre','S√£o Paulo','Rio de Janeiro','Bras√≠lia','Salvador','Curitiba','Belo Horizonte'],
    'IDADE':[14,13,15,16,18,13,14],
    'NOTA':[83.0,59.0,86.0,89.0,98.0,61.0,44.0]
}

dfAlunos = pd.DataFrame(data=dados)

df_media = dfAlunos.groupby('CIDADE')['NOTA'].mean().sort_values()
ax = df_media.plot.bar(rot=0, figsize=(12, 6))

#Adicionando nomenclaturas para melhor visibilidade do gr√°fico
ax.set_title('M√©dia das notas por cidade')
ax.set_xlabel('CIDADES')
ax.set_ylabel('NOTAS')

#Definindo mais ticks no eixo Y
y_ticks = np.arange(0, 101, 10) # Lista de 0 a 100 com intervalos de 10
ax.set_yticks(y_ticks)

print("Gr√°fico:")
plt.show()
```
